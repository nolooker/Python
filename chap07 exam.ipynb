```python
'''
members.csv 파일에 대하여 다음 물음에 답해 보세요.

'int64' 타입의 데이터만 추출해보세요.
'float64' 타입의 데이터만 추출해보세요.
숫자 형식의 타입 모두를 추출해보세요.
컬럼 이름에 'm'이 포함된 모든 컬럼을 추출해 보세요.
컬럼 이름이 'p'로 시작되는 모든 컬럼을 추출해 보세요.
컬럼 이름에 'm' 또는 'n'이 포함된 모든 컬럼을 추출해 보세요.
'name', 'marriage', 'gender' 컬럼 정보를 filter() 메소드를 이용하여 추출해 보세요.

컬럼 재배치를 통하여 다음 컬럼들을 우선적으로 앞 배치하도록 하세요.
'name', 'professional2022', 'professional2023', 'speech2022', 'speech2023'

'''
```

```python
import pandas as pd
```

```python
member = pd.read_csv('./../data/members.csv', encoding='cp949')
member.info()
```

```python
result = member.select_dtypes(include=['int64'])
display(result)
```

```python
result = member.select_dtypes(include=['float64'])
display(result)
```

```python
result = member.select_dtypes(include=['number'])
display(result)
```

```python
result = member.filter(like='m').head()
result
```

```python
result = member.filter(regex='^p').head()
display(result)
```

```python
result = member.filter(regex='(m|n)').head()
result
```

```python
result = member.filter(items=['name', 'marriage','gender']).head()
result
```
