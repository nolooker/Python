```python
import pandas as pd
```

```python
myframe01 = pd.read_csv('./../data/payment01.csv', encoding='cp949')
myframe01
```

```python
myframe02 = pd.read_csv('./../data/payment02.csv', encoding='cp949')
myframe02
```

```python
columnfo01 = myframe01.columns
columnfo02 = myframe02.columns
columnfo01.intersection(columnfo02)
```

```python
columnfo01.difference(columnfo02)
```

```python
columnfo02.difference(columnfo01)
```

```python
concatRow = pd.concat([myframe01, myframe02], axis=0)
concatRow
```

```python
someline = 3
print('%d 번째 라인 정보 출력' % (someline))
third = concatRow.iloc[someline, ]
print(third)
print('\n데이터 유형 :' +str(type(third)))
```

```python
left_frame = pd.DataFrame({
    '지역':['서울','서울','서울','부산','부산'],
    '년도':[2018,2019,2020,2019,2020],
    'data':[10 * idx for idx in range(1,6)]
})

left_frame
```

```python
right_frame = pd.DataFrame(np.arange(12).reshape((6,2)), 
                index=[['부산','부산','서울','서울','서울','서울'],
                        [2019,2018,2018,2018,2019,2020]],
                columns=['컬럼01','컬럼02'])
right_frame             
```

```python
myindex = [['부산','부산','서울','서울','서울','서울'],[2019,2018,2018,2018,2019,2020]]
mycolumn = ['컬럼01','컬럼02']

data = np.arange(12).reshape((6,2))

right_frame = pd.DataFrame(data, index=myindex, columns=mycolumn)
right_frame
```

```python
left_col = ['지역','년도']
result = pd.merge(left_frame, right_frame, left_on=left_col, right_index=True, how='outer')
result
```
