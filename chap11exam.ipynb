```python
def get_range(x):
    return x.between(1000,5000).mean()
```

```python
result = product.groupby(['product'])['price'].agg(get_range)
result
```

```python
def get_range_pct(x, lower, upper) :
    return x.between(lower, upper).mean()
```

```python
LOWER, UPPER = 1500,3000

result = product.groupby(['product'])['price'].agg(get_range_pct,LOWER,UPPER)
result
```

```python
result = product.groupby(['product'])['price'].agg(get_range_pct,lower=LOWER,upper=UPPER)
result.to_frame()
```
